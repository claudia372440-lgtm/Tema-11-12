```python
# Lista con cinco elementos
numeros = [10, 20, 30, 40, 50]

print("La lista creada es:")
print(numeros)
```

```python
# Crear una cadena con tres palabras
frase = "Programacion en Python"

print("La cadena es:")
print(frase)
```

```python
# Contar caracteres
cantidad_caracteres = len(frase)

print("Cantidad de caracteres:")
print(cantidad_caracteres)
```

```python
# Convertir a maysculas y guardar en nueva variable
frase_mayusculas = frase.upper()

print("Cadena en maysculas:")
print(frase_mayusculas)
```

```python
# Crear diccionario con claves alfabticas
diccionario = {
    "A": "Amor",
    "B": "Bondad",
    "C": "Compromiso",
    "D": "Disciplina"
}

print("Diccionario creado:")
print(diccionario)
```

```python
# Agregar nuevo elemento
diccionario["E"] = "Esfuerzo"

print("Diccionario despus de agregar un elemento:")
print(diccionario)
```

```python
# Eliminar la clave "C"
del diccionario["C"]

print("Diccionario despus de eliminar la clave 'C':")
print(diccionario)
```

```python
# Contar elementos
cantidad_elementos = len(diccionario)

print("Cantidad de elementos en el diccionario:")
print(cantidad_elementos)
```
