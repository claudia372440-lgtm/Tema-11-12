Una lista es una estructura de datos que te permite almacenar una colección ordenada de elementos. Piensa en ella como una especie de contenedor o caja donde puedes guardar diferentes tipos de datos, como números, texto, e incluso otras listas.

* Son ordenadas: Los elementos mantienen el orden en el que los agregaste. Cada elemento tiene una posición o índice, comenzando desde 0.

* Son mutables: Puedes cambiar su contenido. Es posible agregar, eliminar o modificar elementos después de haber creado la lista.

* Permiten duplicados: Una lista puede contener el mismo valor más de una vez.

Pueden contener diferentes tipos de datos: En una misma lista puedes mezclar números, cadenas de texto, booleanos (True/False), etc.

```python
dia = "sabado"

match dia:
  case "sabado" | "domingo":
    print("Es fin de semana")
  case "lunes" | "martes" | "miércoles":
    print("Es día laboral")
  case _:
    print("Dia inválido")

```

```python
numero = 16

match numero:
  case n if n < 0:
    print("El número es negativo")
  case n if n > 0:
    print("El número es positivo")
  case n if n == 0:
    print("El número es cero")
```

# Menu

```python
opcion = ""

while opcion != "4":
  print("---MENÚ PRINCPAL---")
  print("1: Ver Conctactos")
  print("2: Agregar un contacto")
  print("3: Eliminar un contacto")
  print("4: Salir")
  opcion = input("Selecciona una opción: ")

  if opcion == "1":
    print("Ver Conctactos")
  elif opcion == "2":
    print("Agregar un contacto")
  elif opcion == "3":
    print("Eliminar un contacto")
  elif opcion == "4":
    print("Salir")
  else:
    print("Opción inválida")
```

```python
contactos = {
    "Hugo Antonio":"5552929290",
    "Joan Gabriel":"8282773790",
    "Mary Zolila":"72818281267"
}

while contactos:
  nombre, telefono = contactos.popitem()
  print("Eliminando contacto: ", nombre, telefono)

print("No hay más contactos")
```

```python
contactos = {}
opcion = ""

while opcion != "3":
  print("---MENÚ PRINCPAL---")
  print("1: Agregar un contacto")
  print("2: Ver contactos")
  print("3: Salir")

  opcion = input("Selecciona una opción: ")

  if opcion == "1":
    nombre = input("Nombre: ")
    telefono = input("Teléfono: ")
    contactos[nombre] = telefono
    print("Contacto agregado")

  elif opcion == "2":
    print("Lista de contactos")
    for nombre, telefono in contactos.items():
      print(nombre, telefono)
  elif opcion == "3":
    print("Salir")
  else:
    print("Opción inválida")



```
